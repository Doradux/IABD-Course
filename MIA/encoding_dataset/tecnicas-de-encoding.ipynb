# Target Encoding

#### ¿Qué es?  
En esta técnica, cada categoría de una variable se reemplaza por una métrica derivada del objetivo (por ejemplo, la media del objetivo para cada categoría).
Es especialmente útil en problemas de clasificación o regresión.
#### Ventajas:
- Reduce la dimensionalidad: A diferencia de One-Hot Encoding, no genera nuevas columnas.
- Captura relaciones directas entre las categorías y el objetivo.
- Funciona bien con datos categóricos de alta cardinalidad.
#### Desventajas:
- Puede causar fuga de datos si no se hace correctamente (el objetivo futuro se filtra al modelo durante el entrenamiento).
- Es sensible a categorías con pocos datos (puede introducir ruido).

## Ingreso de personas Adultas

```python
adult_census_df = pd.read_csv('/kaggle/input/adult-census-income-dataset/adult.csv')
x = adult_census_df
x['income'] = x['income'].map({'<=50K': 0, '>50K': 1})
x.head()
```

```python
categorical_columns = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']
y = x[categorical_columns]
y.head()
```

```python
# Aplicar Target Encoding a cada columna categórica
for col in categorical_columns:
    mean_income_per_category = x.groupby(col)['income'].mean()
    x[col] = x[col].map(mean_income_per_category)

# Mostrar el dataset con Target Encoding aplicado
x = x[categorical_columns + ['income']]
x.head()
```

#### Significado

Cada valor numérico en las nuevas columnas refleja la media del ingreso (en este caso, 0 o 1) para cada categoría. Por ejemplo, el valor 0.104031 en la columna workclass significa que la media de income para esa categoría específica en la columna workclass es aproximadamente 0.104031.

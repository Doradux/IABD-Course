# Target Encoding

#### ¬øQu√© es?  
En esta t√©cnica, cada categor√≠a de una variable se reemplaza por una m√©trica derivada del objetivo (por ejemplo, la media del objetivo para cada categor√≠a).
Es especialmente √∫til en problemas de clasificaci√≥n o regresi√≥n.
#### Ventajas:
- Reduce la dimensionalidad: A diferencia de One-Hot Encoding, no genera nuevas columnas.
- Captura relaciones directas entre las categor√≠as y el objetivo.
- Funciona bien con datos categ√≥ricos de alta cardinalidad.
#### Desventajas:
- Puede causar fuga de datos si no se hace correctamente (el objetivo futuro se filtra al modelo durante el entrenamiento).
- Es sensible a categor√≠as con pocos datos (puede introducir ruido).

## Ingreso de personas Adultas

```python
adult_census_df = pd.read_csv('/kaggle/input/adult-census-income-dataset/adult.csv')
x = adult_census_df
x['income'] = x['income'].map({'<=50K': 0, '>50K': 1})
x.head()
```

```python
categorical_columns = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']
y = x[categorical_columns]
y.head()
```

```python
# Aplicar Target Encoding a cada columna categ√≥rica
for col in categorical_columns:
    mean_income_per_category = x.groupby(col)['income'].mean()
    x[col] = x[col].map(mean_income_per_category)

# Mostrar el dataset con Target Encoding aplicado
x = x[categorical_columns + ['income']]
x.head()
```

#### Significado

Cada valor num√©rico en las nuevas columnas refleja la media del ingreso (en este caso, 0 o 1) para cada categor√≠a. Por ejemplo, el valor 0.104031 en la columna workclass significa que la media de income para esa categor√≠a espec√≠fica en la columna workclass es aproximadamente 0.104031.

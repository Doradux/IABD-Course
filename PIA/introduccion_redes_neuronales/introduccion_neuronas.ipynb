Imports

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
import random
import tensorflow as tf
```

Datos cancer de mama

```python
breast_cancer = load_breast_cancer()
x = breast_cancer.data
y = breast_cancer.target
```

Fila 56 como muestra el libro

```python
#Para que los datos no se muestren con notación científica
np.set_printoptions(suppress=True)

print(x[56],y[56])
```

Mostrar fila 204

```python
print(x[204], y[204])
```

Crear modelo neurona

```python
np.random.seed(5)
tf.random.set_seed(5)
random.seed(5)

model = Sequential()
model.add(Dense(30, activation='relu', input_dim=30))
model.add(Dense(60, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(60, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy')
```

Compilar y entrenar modelo

```python
model.fit(x, y, epochs=40)
```

Predicciones fila 56 y 204

```python
print(f'56: {y[56]} ==> {model.predict(np.array([x[56]]))}')
print(f'204: {y[204]} ==> {model.predict(np.array([x[204]]))}')
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.DataFrame(data=breast_cancer.data, columns=breast_cancer.feature_names)
df['target'] = breast_cancer.target
```

```python
df.head(3)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
df = df.drop(["mean perimeter", "mean area", "worst radius", "worst perimeter", "worst area", "area error", "perimeter error", "radius error", "worst concave points", "mean concave points", ], axis=1)
```

```python
len(df.columns)
```

```python
cols_existentes = [col for col in breast_cancer.feature_names if col in df.columns]

x = df[cols_existentes].to_numpy()
y = df['target'].to_numpy()
```

```python
# Dividimos los datos en conjuntos de entrenamiento (80%) y prueba (20%)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Inicializamos el generador de números para tener resultados reproducibles
np.random.seed(5)
tf.random.set_seed(5)
random.seed(5)

#Creamos el modelo (uno más pequeño que el anterior)
model=Sequential()
model.add(Dense(2, activation='selu',input_dim=20))
model.add(Dense(20, activation='selu'))
model.add(Dense(10, activation='selu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy')


model.fit(x_train, y_train,validation_data=(x_test,y_test),epochs=40) # El batch_size por defecto es 32
```
